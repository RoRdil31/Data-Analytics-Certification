## 57. Map (Series만 가능)

## 59. ca값을 이용해서 나이구간을 as 컬럼으로 정의. 0~9:0, 10~19:10,... 각 구간의 빈도수 출력.

```python
# def code(age):
#     return age//10 * 10

df.AgeState = df.Customer_Age.apply(lambda x : x//10*10)
df.AgeState.value_counts().sort_index()
```

```python
# 60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라
# df.newEduLevel = df['Education_Level'].str.contains('Graduate').apply(lambda x : 1 if x else 0)
# df.newEduLevel.value_counts()
```

```python
# 61. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라
df.newLimit = (df.Credit_Limit >= 4500).apply(lambda x : 1 if x else 0)
df.newLimit.value_counts()
```

## 67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
import datetime
def code(x):
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(code)
df.head(3)
```

## 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

Ans = df
```

## 79. 시간이 연속적으로 존재하며 결측치가 없는지 확인하라

```python
# 시간을 차분했을 경우 첫 값은 nan, 이후 모든 차분값이 동일하면 연속이라 판단한다.
check = len(df['(년-월-일:시)'].diff().unique())
if check ==2:
    Ans =True
else:
    Ans = False
print(df['(년-월-일:시)'].diff())
print(Ans)
```
